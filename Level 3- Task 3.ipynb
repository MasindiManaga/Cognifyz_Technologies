#  Data Visualization

```python
#import necessary libraries
import numpy as np
import pandas as pd
```

```python
#load the dataset
df= pd.read_csv("C:/Users/manag/Cognifyz Projects/Restaurant.csv")
df.head()
```

```python
#represent the distribution of ratings using histogram
import matplotlib.pyplot as plt

plt.hist(df['Aggregate rating'], bins=5, color='blue', edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
#represent the distribution of ratings using bar plot
counts = df['Aggregate rating'].value_counts().sort_index()

counts.plot(kind='bar', color='skyblue')
plt.xlabel('Aggregate rating')
plt.ylabel('Count')
plt.title('Distribution of ratings')
plt.show()
```

```python
#represent the distribution of ratings using Pie chart
counts = df['Aggregate rating'].value_counts().sort_index().head(20)

counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, textprops={'fontsize':7, 'rotation_mode': 'anchor', 'ha': 'center'}, cmap='viridis')
plt.title('Distribution of Ratings')
plt.tight_layout()
plt.show()
```

```python
#Comapring average ratings of different cities
city_ratings = df.groupby("City")["Aggregate rating"].mean().head(10)
city_ratings
```

```python
#Visualize average rating of different cities

city_ratings.plot(kind='bar', color='pink')
plt.xlabel('City')
plt.ylabel('Average rating')
plt.title('Average ratings of cities')
plt.show()
```

```python
#Average rating of  Restaurants/ restaurant name
restaurant_ratings = df.groupby("Restaurant Name")["Aggregate rating"].mean().head(10)
restaurant_ratings

restaurant_ratings.plot(kind='bar', color='red')
plt.xlabel('Restaurant Name')
plt.ylabel('Average rating')
plt.title('Average ratings of Restaurants')
plt.show()
```

```python
#Average rating of  Locality
Locality_ratings = df.groupby("Locality")["Aggregate rating"].mean().head(10)
Locality_ratings

Locality_ratings.plot(kind='bar', color='orange')
plt.xlabel('Locality')
plt.ylabel('Average rating')
plt.title('Average ratings of Locality')
plt.show()
```

```python
#Average rating of  restaurants with or without table booking
booking_ratings = df.groupby("Has Table booking")["Aggregate rating"].mean().head(10)
booking_ratings

booking_ratings.plot(kind='bar', color='yellow')
plt.xlabel('Table Booking')
plt.ylabel('Average rating')
plt.title('Average ratings of restaurants with or without table booking')
plt.show()
```

```python
#Average rating of  restaurants with or without online delivery
delivery_ratings = df.groupby("Has Online delivery")["Aggregate rating"].mean().head(10)
delivery_ratings

delivery_ratings.plot(kind='bar', color='green')
plt.xlabel('Online Delivery')
plt.ylabel('Average rating')
plt.title('Average ratings of restaurants with or without online delivery')
plt.show()
```
